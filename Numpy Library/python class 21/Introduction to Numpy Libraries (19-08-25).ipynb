### Indexing in Arrays:

```python
arr6 = np.arange(25).reshape(5,5)
arr6
```

```python
print(arr6[2,3])
```

```python
arr7 = np.arange(64).reshape(2,4,8)
arr7
```

```python
print(arr7[1,2,3])
```

### Slicing in Arrays: 

```python
print(arr6[2:4, 1:4])
```

```python
print(arr6[:, 1:3])
```

```python
print(arr6[2:4,:])
```

```python
print(arr6[:, 4:])
```

```python
print(arr6[2:3, 2:3])
```

```python
print(arr6[:,::2])
```

```python
print(arr6[::2,::2])
```

```python
print(arr6[::4,::4])
```

### Image processing with Numpy and PIL:

```python
from PIL import Image # Importing PIL image library

image = Image.open("wallpaper.png") # importing / opening image       
image # displaying image
```

```python
image_array = np.array(image).astype('float32') # converting image into array with float type values
```

```python
image_array # displaying colored imaage pixls values as array
```

```python
image_array.shape # displaying image size in terms of pixls after converting into array, image consist of RCNC = Rows, Columns and No.of channels
```

```python
image = Image.open("wallpaper.png").convert("L") # Convert colored image into grayscale
image
```

```python
image_array = np.array(image).astype('float32') # converting grayscale image into array
```

```python
image_array # displaying grayscale imaage pixls values as array
```

```python
image_array.shape # displaying grayscale image size in terms of pixls after converting into array, no RGB or RGBA color channels in grayscale image
```
