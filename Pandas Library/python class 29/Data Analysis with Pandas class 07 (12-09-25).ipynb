```python
import pandas as pd
```

```python
df1 = pd.read_excel("test.xlsx", sheet_name=0)
df2 = pd.read_excel("test.xlsx", sheet_name=1)
df3 = pd.read_excel("test.xlsx", sheet_name=2)
```

```python
 pd.concat([df2,df1], axis=0, ignore_index=True)
```

```python
 pd.concat([df2,df1], axis=1)
```

```python
pd.merge(df3,df1)
# inner merge, outer merge, left merge, right merge
# by default it merges on overlapping column
# by default it used inner join to merge
# inner join select only match record from both overlapping columns
```

```python
df1.columns = ['sales_man_id', 'name', 'sales_man_city', 'product_id', 'manager']
```

```python
df1
```

```python
pd.merge(df3,df1,left_on="sales_man_name", right_on="name", how="innner").drop("name", axis=1)
# if there is no overlapping column but have columns with same data, we can explicity mention lefton and right on
```

```python
pd.merge(df1,df3,left_on="name", right_on="sales_man_name", how="outer").drop("sales_man_name", axis=1)
```

```python
pd.merge(df1,df3,left_on="name", right_on="sales_man_name", how="left").drop("sales_man_name", axis=1)
```

```python
pd.merge(df1,df3,left_on="name", right_on="sales_man_name", how="right").drop("sales_man_name", axis=1)
```
