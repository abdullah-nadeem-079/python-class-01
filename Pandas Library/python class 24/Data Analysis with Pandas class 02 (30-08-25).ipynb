```python
import pandas as pd
```

```python
std = pd.read_csv("student.csv")
std
```

```python
std.apply(lambda x: sum(x[["Python Marks", "Numpy Marks", "Pandas Marks", "Stats Marks", "ML Marks"]]),axis = 1)
```

```python
std["Total_obt"] = std["Python Marks"] + std["Numpy Marks"] +std["Pandas Marks"] + std["Stats Marks"] + std["ML Marks"]
std
```

```python
std["Total"] = 500
std
```

```python
std["Percentage"] = std["Total_obt"] / std["Total"] * 100
std
```

```python
std["Grade"] = std.Percentage.apply(lambda x: "A+" if x>=90
                                       else"A" if x >= 80
                                       else"B" if x >= 70
                                       else"C" if x >= 60
                                       else"D" if x >= 50
                                       else "F")
std
```

```python
del std["Roll Number"]
std
```
