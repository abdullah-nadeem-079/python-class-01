* if data is categorical we will use mode.

### Data Transformation:

##### Removing Duplicates

```python
emp_01 = pd.read_excel("employee 01.xlsx")
```

```python
emp_01.duplicated().sum()
```

```python
emp_02 = pd.read_excel("employee 02.xlsx")
```

```python
emp_02
```

```python
emp_02 = pd.read_excel("employee 02.xlsx").sample(26)
```

```python
emp_02.duplicated()
```

```python
emp_02.duplicated(subset = ["Emp_ID", "Age"]).sum()
```

```python
emp_02[emp_02.duplicated(subset = ["Emp_ID", "Age"])]
```

```python
emp_02.drop_duplicates(subset = ["Emp_ID", "Designation", "Department", "Age", "Salary"], inplace = True)
```

```python
emp_02.shape
```

```python
emp_02.info()
```

```python
emp_02.isna().sum()
```

```python
emp_02.fillna({"Department" : emp_02.Department.mode()[0], "Age" : emp_02.Age.mean(), "salary" : emp_02.Salary.median()}, inplace = True)
```

```python
emp_02.info()
```

### Data Transformation functions:
* .transform()

```python
df = pd.DataFrame(["cat", "cow", "dog", "lion", "elephant", "dog", "monkey", "cat", "lion", "elephant"], columns = ["animals"])
df
```

```python
dictionary = {
    "cat" : "pet",
    "cow" : "pet",
    "dog" : "pet",
    "lion" : "wild",
    "elephant" : "wild",
    "dog" : "pet",
    "monkey" : "wild",
    "cat" : "pet",
    "lion" : "wild",
    "elephant" : "wild"
}
```

```python
df["type"] = df.animals.map(dictionary)
df
```
