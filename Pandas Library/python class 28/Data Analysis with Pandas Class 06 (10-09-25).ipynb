### Detecting and Filtering Outliers:

```python
import seaborn as sns
```

```python
sns.boxplot(data.Age)
```

```python
sns.boxplot(data.Salary)
```

```python
data.Age.fillna(data.Age.mean(), inplace=True)
data.Salary.fillna(data.Salary.median(), inplace=True)
```

```python
import numpy as np
```

```python
def outlier_detector(col):
    q1 = np.percentile(col,25)
    q3 = np.percentile(col,75)
    IQR = q3-q1
    lower_boundary = q1-1.5*IQR
    upper_boundary = q3+1.5*IQR
    return lower_boundary,upper_boundary
```

```python
lb,ub = outlier_detector(data.Salary)
```

```python
lb,ub
```

```python
data[(data.Salary<lb) | (data.Salary>ub)]
```

```python
data[(data.Salary<outlier_detector(data.Salary)[0]) | (data.Salary>outlier_detector(data.Salary)[1])]
```

```python
sns.boxplot(data[(data.Salary>=lb) & (data.Salary<=ub)]["Salary"])
```
