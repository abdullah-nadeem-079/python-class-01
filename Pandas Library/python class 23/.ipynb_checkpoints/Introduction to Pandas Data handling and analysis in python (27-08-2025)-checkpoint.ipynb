```python
emp["Tax%"] = 15
emp
```

```python
def decide_tax_percentage(sal):
    if sal >= 200000:
        return 20
    elif sal >= 150000:
        15
    elif sal >= 100000:
        10
    else:
        return 5
```

```python
emp.Salary.apply(decide_tax_percentage)
```

```python
emp["Tax%"] = emp.Salary.apply(decide_tax_percentage)
emp
```

```python
emp["tax%"] = emp.Salary.apply(lambda x: 20 if x >= 200000 else 15 if x >= 150000 else 10 if x >= 100000 else 5)
```

```python
emp["tax_amt"] = emp.Salary.apply(lambda x: x * .20 if x >= 200000 else x * .15 if x >= 150000 else x * .10 if x >= 100000 else x * .05)
```

```python
emp
```

```python
emp["net_salary"] = emp["Salary"] - emp["tax_amt"]
```
